# ‚è© Lenguajes de Programaci√≥n


## Que es un Lenguaje de Programacion?

Un lenguaje de programaci√≥n es un **conjunto de reglas** e **instrucciones** que permiten a los programadores comunicarse con una computadora. Los programadores usan estos lenguajes para crear programas y aplicaciones.

Diferencia entre Algoritmo y un Programa
Un **algoritmo** es un conjunto de pasos organizados de manera l√≥gica para resolver un problema o realizar una tarea.

Por otro lado, un **programa** es la implementaci√≥n de un algoritmo en un **lenguaje de programaci√≥n** para que una computadora pueda ejecutarlo. Es decir, el algoritmo es la idea y el programa es la versi√≥n escrita en un lenguaje que la computadora entiende (como Python, C++, Java, etc.).


## Lenguaje m√°quina vs. lenguaje de alto nivel
El lenguaje de m√°quina, tambi√©n conocido como c√≥digo m√°quina, es el lenguaje m√°s b√°sico que una computadora puede entender. Consiste en una serie de n√∫meros binarios (ceros y unos) que representan instrucciones directamente ejecutables por el hardware. Es el lenguaje en el que se traducen todos los dem√°s lenguajes de programaci√≥n para que la computadora pueda llevar a cabo las tareas.
Caracter√≠sticas principales:

* Nivel m√°s bajo:
Es el nivel m√°s cercano al hardware de la computadora.
* C√≥digo binario:
Est√° compuesto exclusivamente por ceros y unos, que representan instrucciones y datos.
* Espec√≠fico de la arquitectura:
El lenguaje m√°quina es diferente para cada tipo de procesador o arquitectura de computadora.
* Dif√≠cil de entender para humanos:
Es muy dif√≠cil de leer y escribir directamente por los humanos, ya que no es intuitivo.
* Base de la programaci√≥n:
Es el lenguaje fundamental en el que se basan todos los dem√°s lenguajes de programaci√≥n.

### ¬øC√≥mo funciona?
Un programa escrito en un lenguaje de alto nivel (como Python, Java, etc.) debe ser traducido al lenguaje m√°quina para que la computadora pueda ejecutarlo. Este proceso de traducci√≥n se realiza mediante **compiladores o int√©rpretes**. Una vez traducido, el c√≥digo m√°quina se carga en la memoria de la computadora y el procesador lo ejecuta.

Ejemplos de instrucciones en lenguaje m√°quina:
* Instrucciones de transferencia de datos: Mover datos de una ubicaci√≥n a otra en la memoria.
* Instrucciones aritm√©ticas: Sumar, restar, multiplicar o dividir datos.
* Instrucciones de control: Tomar decisiones basadas en ciertas condiciones (saltos, bucles).

En resumen, el lenguaje m√°quina es el lenguaje nativo de las computadoras, fundamental para su funcionamiento, aunque su uso directo por parte de los humanos es limitado debido a su complejidad.

Ejemplos de Lenguajes de Bajo Nivel:
1. Lenguaje ensamblador: Es un lenguaje simb√≥lico que representa las instrucciones del c√≥digo m√°quina de una manera m√°s legible para los humanos. Cada arquitectura de procesador tiene su propio lenguaje ensamblador.
Por ejemplo, el lenguaje ensamblador x86 se utiliza en procesadores Intel y compatibles, mientras que el ensamblador MIPS se utiliza en sistemas MIPS.
Otros ejemplos de lenguajes de bajo nivel:
2. Lenguajes HDL (Hardware Description Languages):
Se utilizan para describir el dise√±o de circuitos electr√≥nicos y hardware, como procesadores y chips.
3. Ensamblador ARM:
Es popular en sistemas embebidos, dispositivos m√≥viles y sistemas basados en Linux, con una arquitectura RISC que facilita la ejecuci√≥n r√°pida de instrucciones


## Lenguajes de Programaci√≥n Compilados o Interpretados

* Compilados: Se traduce todo el codigo a lenguaje de maquina antes de ejecutarlo.
* Interpretados: Se traduce a lenguaje de maquina a medida que se va ejecutando el codigo.

Los lenguajes de programaci√≥n **compilados y los interpretados** se diferencian en c√≥mo se ejecuta el c√≥digo. Los compilados son m√°s r√°pidos y eficientes, pero los interpretados son m√°s flexibles.

Los compiladores y los int√©rpretes son programas que convierten el c√≥digo legible por humanos en c√≥digo m√°quina legible por computadoras.

Los lenguajes interpretados son fundamentales en el desarrollo web y otras √°reas de la programaci√≥n. Esto se debe a su flexibilidad, facilidad de uso y capacidad de adaptaci√≥n a m√∫ltiples plataformas.

## Sintaxis

La sintaxis de un lenguaje de programaci√≥n es el conjunto de reglas que determinan c√≥mo se escribe el c√≥digo. Es similar a las reglas gramaticales de un idioma.
* La sintaxis es importante porque permite que el c√≥digo sea legible para los programadores y las m√°quinas.
* Cada lenguaje de programaci√≥n tiene su propia sintaxis, aunque algunos comparten elementos comunes.

###Errores de sintaxis
Si un programador no sigue la sintaxis correcta, el c√≥digo puede tener errores de sintaxis. Estos errores impiden que el programa se compile o se ejecute.
###Herramientas para la sintaxis
El resaltado de sintaxis, carateres de finalizacion de instrucciones y el estilo de sangr√≠a son herramientas que ayudan a los programadores a reconocer los elementos del c√≥digo.

## Tipado en los Lenguajes de Programaci√≥n

### Fuertemente Tipado
Un lenguaje fuertemente tipado es un lenguaje de programaci√≥n que impone reglas estrictas sobre los tipos de datos y las operaciones entre variables. Estas reglas est√°n dise√±adas para evitar errores de programaci√≥n.

El compilador de un lenguaje fuertemente tipado:
*	Detecta errores relacionados con tipos de variable incorrectos
*	Proh√≠be operaciones entre variables de distintos tipos
*	Garantiza la conformidad con la tipificaci√≥n de datos

El tipado fuerte es recomendable porque:

*	Exige castear o comprobar el tipo de cada variable
*	Ayuda al programador a filtrar bugs relacionados a los tipos
*	Es m√°s seguro, porque no permite hacer operaciones con variable de distintos tipos.

En contraste, un lenguaje d√©bilmente tipado es m√°s flexible y permite usar variables y valores sin especificar sus tipos. Por ejemplo, JavaScript es un lenguaje d√©bilmente tipado.

### D√©bilmente Tipado
Un lenguaje de programaci√≥n d√©bilmente tipado es aquel que no requiere especificar el tipo de cada variable. Esto permite asignar valores de distintos tipos a una misma variable, sin necesidad de realizar conversiones expl√≠citas.
Caracter√≠sticas
*	Permiten cambiar de tipos de datos "al vuelo"
*	Permiten usar y manipular variables y valores sin especificar sus tipos
*	Pueden hacer que el c√≥digo sea m√°s conciso y sencillo
*	Pueden mejorar la flexibilidad y creatividad

Ejemplos JavaScript, PHP, Perl.

### Ventajas y desventajas
Entre sus ventajas, se encuentran la posibilidad de reducir la cantidad de c√≥digo que se debe escribir y mantener. Sin embargo, tambi√©n pueden provocar comportamientos inesperados y dificultar la depuraci√≥n.

### C√≥mo funcionan
En los lenguajes d√©bilmente tipados, el int√©rprete o compilador permite la coerci√≥n impl√≠cita, lo que significa que autom√°ticamente convierte valores entre diferentes tipos de datos.

## Nombres de las Variables
En Leguajes de Programaci√≥n estructurados debemos tener algunas reglas de sintaxis en cuenta.

Si se desea nombrar una variable, se deben seguir las siguientes reglas:

**Nombres de Variables:**
*   Se recomienda mas de 4 caracteres y menos de 20, salvo para contadores de uso temporal de 1 (un) car√°cter como i, j, n, x etc..
*   El nombre de la variable debe de estar compuesto por MAY√öSCULAS, min√∫sculas, d√≠gitos, y el car√°cter _ (guion bajo)
*   El nombre de la variable debe comenzar con una letra;
*   En la mayor√≠a de los lenguajes el nombre de la variable NO puede empezar con un n√∫meros o s√≠mbolo. (No se permite comenzar con 0-9 ni $, &, !)
*   El car√°cter guion bajo es considerado una letra;
*   No se permiten espacios
*   Las may√∫sculas y min√∫sculas se tratan de forma distinta. (distigue entre Mayuscula y minuscula)
*   El nombre de las variables no puede ser igual a alguna de las palabras reservadas.

##CONVENCIONES
1. **snake_case** para nombres de variables:
La convenci√≥n m√°s com√∫n para nombres de variables en Python es utilizar snake_case, donde las palabras se separan por guiones bajos y se escriben en min√∫sculas.

  *Por ejemplo: mi_variable, precio_total. *
2. **lowerCamelCase** para nombres de m√©todos:
Aunque menos com√∫n para variables, lowerCamelCase (donde la primera letra de cada palabra, excepto la primera, es may√∫scula) se utiliza para nombrar m√©todos.
  *Por ejemplo: miVariable, precioTotal. *

3. Nombres descriptivos:
Los nombres de las variables deben ser lo m√°s descriptivos posible para que el c√≥digo sea f√°cil de entender.

4. Evitar nombres de una sola letra:
Excepto en casos espec√≠ficos como contadores de bucles, se recomienda evitar nombres de variables de una sola letra, como x, y, etc.

5. Constantes en may√∫sculas:
Para variables que se consideran constantes (es decir, que no deber√≠an cambiar), se utiliza la convenci√≥n SNAKE_CASE (todo en may√∫sculas y separado por guiones bajos).
  
  *Por ejemplo: PI, GRAVEDAD, COLOR_BK*

## Condicional CASE (SEG√öN ‚Äì SWITCH-CASE / MATCH-CASE)

**Ejemplo**: Selecci√≥n de un men√∫ de bebidas.

image.png

### En PSeInt
    Algoritmo Menu
    Escribir "Seleccione una opci√≥n:"
    Escribir "1. Caf√© negro"
    Escribir "2. Caf√© con leche"
    Escribir "3. T√©"
    Leer opcion

    Seg√∫n opcion Hacer
        1: Escribir "Usted eligi√≥ Caf√© negro."
        2: Escribir "Usted eligi√≥ Caf√© con leche."
        3: Escribir "Usted eligi√≥ T√©."
        De Otro Modo:
            Escribir "Opci√≥n no v√°lida."
    FinSeg√∫n
    Fin Algoritmo


### En Python

````python
opcion = int(input("Seleccione una opci√≥n:\n1. Caf√© negro\n2. Caf√© con leche\n3. T√©\n"))

match opcion:
    case 1:
        print("Usted eligi√≥ Caf√© negro.")
    case 2:
        print("Usted eligi√≥ Caf√© con leche.")
    case 3:
        print("Usted eligi√≥ T√©.")
    case _:
        print("Opci√≥n no v√°lida.")
````

# Repetitivas

## Repetir Mientras (WHILE)

Se usa cuando no sabemos cu√°ntas veces se repetir√° y la condici√≥n se eval√∫a al principio.

**Ejemplo**: Pedir una clave hasta que sea la correcta (1234).

### DFD

image.png

### En PSeInt


    Algoritmo IngresoSistema
      Escribir "Ingrese la clave:"
      Leer clave
    Mientras clave <> "1234" Hacer
      Escribir "Clave incorrecta. Intente nuevamente:"
      Leer clave
    FinMientras
     Escribir "¬°Clave correcta!"
    Fin Algoritmo

### En Python

````python
clave = input("Ingrese la clave: ")

while clave != "1234":
    clave = input("Clave incorrecta. Intente nuevamente: ")

print("¬°Clave correcta!")
````

## Repetir Hasta Que (DO - WHILE)

Se usa cuando queremos que el bloque se ejecute al menos una vez, y la condici√≥n se eval√∫e al final.

En Python y C++ no existe como estructura directa, pero se simula.

**Ejemplo**: Pedir texto hasta que el usuario escriba "salir".

### DFD

image.png


### En PSeInt

    Repetir
      Escribir "Escriba algo (o 'salir' para terminar):"
      Leer texto
    Hasta Que texto = "salir"


### En Python

````python
texto = str

while texto != "salir":
    texto = input("Escriba algo (o 'salir' para terminar): ")
````
